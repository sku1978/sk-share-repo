```python
with open('my_template.json') as f:
    d = json.load(f)
```

```python
d['resources']
```

```python
import re
def get_resource_name(value):
  result=re.search("\[concat\(parameters\(\'factoryName\'\), \'(.*)\'\)\]", value)
  return result.group(1)
```

```python
resources = json_normalize(d['resources'])
resource_1=resources[['name', 'type', 'properties.type', 'dependsOn']].copy()
resource_1['name']=resource_1['name'].apply(get_resource_name)
#resource_1['name']=resource_1['name'].apply(lambda x: literal_eval(x.replace('[concat', '["concat').replace(')]', ')"]')))
#resource_1['dependsOn'][157]
resource_1
```

```python
resource_1.explode('dependsOn')
```

```python
import numpy as np
pd.DataFrame({'name':np.repeat(resource_1.name.values, resource_1.dependsOn.str.len()),
                        'dependsOn':np.concatenate(resource_1.dependsOn.values)})
```
